```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.Series(np.random.rand(10),index=range(2015, 2025))
```

```python
data
```

```python
# Line plot
plt.figure(figsize = (8, 6))
data.plot()
plt.title("Random Data")
plt.xlabel("Year")
plt.ylabel("Value")
```

```python
# Bar chart
plt.figure(figsize = (8, 6))
data.plot(kind="bar")
plt.title("Random Data")
plt.xlabel("Year")
plt.ylabel("Value")
```

```python
# Histogram
plt.figure(figsize = (8, 6))
data.plot.hist(bins=5,edgecolor="black")
plt.title("Random Data")
plt.xlabel("Year")
plt.ylabel("Value")
```

```python
# Pie CHart
plt.figure(figsize = (8, 6))
data.nlargest(5).plot(kind = "pie", autopct="%1.1f%%")
plt.title("Top 5 Values")
plt.show()
```
