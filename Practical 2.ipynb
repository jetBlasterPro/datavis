```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = {"City": ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"], "Population" : [8.4, 4, 2.7, 2.3, 1.7], "Year" : [2020, 2020, 2020, 2020, 2020]}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Barchart of population
plt.figure(figsize = (8, 6))
sns.barplot(x = "City", y = "Population", data = df)
plt.title("City Population")
plt.xlabel("City")
plt.ylabel("Population (million)")
plt.show()
```

```python
# Line plot of city population
plt.figure(figsize = (8, 6))
plt.plot(df["City"], df["Population"], marker="o")
plt.title("City Population")
plt.xlabel("City")
plt.ylabel("Population (million)")
plt.show()
```

```python
# Line plot of city population
plt.figure(figsize = (8, 6))
plt.pie(df["Population"], labels = df["City"], autopct ="%1.1f%%")
plt.title("City Population Distrubution")
plt.show()
```

```python
# Line plot of city population
plt.figure(figsize = (8, 6))
plt.scatter(df["Year"], df["Population"])
plt.title("City Population Vs Year")
plt.xlabel("Year")
plt.ylabel("Population (million)")
plt.show()
```
